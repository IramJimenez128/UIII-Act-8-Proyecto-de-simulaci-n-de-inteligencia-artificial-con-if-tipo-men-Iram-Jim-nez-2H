#### Problema director y batuta

```python
# SIMULACIÓN DE IA: ASISTENTE TÉCNICO DE BATUTA
# Control de flujo mediante ciclos y condicionales

# Variable para controlar la permanencia en el menú
continuar = "si"

print("--- SISTEMA IA: BATUTA INTELIGENTE V1.0 ---")
print("IA: Saludos, Maestro. Estoy lista para seguir sus movimientos.")

# Ciclo principal para mantener el programa activo [1, 4]
while continuar == "si":
    print("\n--- MENÚ DE CONTROL DE LA BATUTA ---")
    print("1. Iniciar Compás (Ritmo)")
    print("2. Ajustar Intensidad (Dinámica)")
    print("3. Analizar Final de Pieza")
    print("4. Apagar Sistema IA")

    # Captura de datos del usuario [5, 6]
    opcion = input("\nDirector, elija una acción (1-4): ")

    # Rama 1: Análisis de Ritmo (Operadores Relacionales) [7, 8]
    if opcion == "1":
        print("IA: Detectando velocidad de la batuta...")
        bpm = int(input("IA: ¿Cuántos pulsos por minuto (BPM) marca su batuta?: "))

        if bpm < 60:
            # Uso de operador relacional menor que [8, 9]
            print("IA DIAGNÓSTICO: Ritmo 'Adagio'. Los músicos tocarán lento y solemne.")
        elif bpm >= 60 and bpm <= 120:
            # Operador lógico 'and' para evaluar un rango [8, 10]
            print("IA DIAGNÓSTICO: Ritmo 'Moderato'. Ejecución estándar confirmada.")
        else:
            print("IA DIAGNÓSTICO: Ritmo 'Allegro'. ¡Cuidado con la precisión de las cuerdas!")

    # Rama 2: Ajuste de Intensidad (Operadores Aritméticos) [11, 12]
    elif opcion == "2":
        print("IA: Evaluando la fuerza del movimiento...")
        nivel_fuerza = float(input("IA: Ingrese la amplitud del movimiento (1-10): "))

        if nivel_fuerza > 8:
            print("IA RECOMENDACIÓN: 'Fortissimo'. Indique a los metales que tengan cuidado.")
        elif nivel_fuerza > 4:
            print("IA RECOMENDACIÓN: 'Mezzoforte'. Intensidad equilibrada.")
        else:
            print("IA RECOMENDACIÓN: 'Pianissimo'. Movimiento sutil para un cierre delicado.")

    # Rama 3: Final de Pieza (Lógica de decisión compleja) [3, 8]
    elif opcion == "3":
        cierre = input("IA: ¿El movimiento de la batuta fue seco y tajante? (si/no): ")

        if cierre == "si":
            print("IA: Procesando corte final... ¡Excelente cierre, Maestro!")
        else:
            print("IA: El movimiento es fluido. La orquesta mantendrá la nota (Sostenuto).")

    # Rama 4: Salir del programa [1, 13]
    elif opcion == "4":
        print("IA: Desconectando sensores de la batuta. Hasta pronto, Maestro.")
        continuar = "no" # Cambia el valor para romper el ciclo while [1, 2]

    # Manejo de opciones inválidas en el menú [3]
    else:
        print("IA: Error. Por favor elija una opción válida entre 1 y 4.")

print("\n--- SESIÓN FINALIZADA ---")
print(" Programa elaborado por Iram Jimenez 2H 0079")
```
